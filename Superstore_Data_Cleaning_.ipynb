# **Product Rating Null Value Handing**

```python
df['Product Rating'].isnull().sum()
```

```python
df['Product Rating']= df.groupby(['Category','Sub-Category','Segment'])['Product Rating'].transform(
lambda x: x.fillna(x.mean())
)
```

```python
df['Product Rating'].isnull().sum()
```

## **Order Priority Missing Values**

```python
df['Order Priority'].isnull().sum()
```

```python
df['Order Priority']= df.groupby('Segment')['Order Priority'].transform(
lambda x: x.fillna(x.mode()[0])
)
```

```python
df['Order Priority'].isnull().sum()
```

# **Shipping Cost Missing Value handling**

```python
df['Shipping Cost'].isnull().sum()
```

```python
df['Shipping Cost'] = df.groupby(['State','Ship Mode'])['Shipping Cost'].transform(
lambda x: x.fillna(x.median())
)
```

```python
df['Shipping Cost'].isnull().sum()
```

# **Marketing Channel missing value handling**

```python
df['Marketing Channel'].isnull().sum()
```

```python
df['Marketing Channel'] = df.groupby(['Segment', 'State','Category','Sub-Category'])['Marketing Channel'].transform(
    lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'Unknown')
)
```

```python
df['Marketing Channel'].isnull().sum()
```

# **Fixing Appropriate Datatype for each Column**

```python
check_data_type = df.dtypes
check_data_type
```

```python
df['Order ID'] = df['Order ID'].astype(str)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%m/%d/%Y')
df['Ship Mode'] = df['Ship Mode'].astype(str)
df['Customer ID'] = df['Customer ID'].astype(str)
df['Customer Name'] = df['Customer Name'].astype(str)
df['Segment'] = df['Segment'].astype(str)
df['Country'] = df['Country'].astype('category')
df['City'] = df['City'].astype('category')
df['State'] = df['State'].astype('category')
df['Postal Code'] = df['Postal Code'].astype(str)
df['Region'] = df['Region'].astype('category')
df['Product ID'] = df['Product ID'].astype(str)
df['Product Name'] = df['Product Name'].astype(str)
df['Category'] = df['Category'].astype('category')
df['Sub-Category'] = df['Sub-Category'].astype('category')
df['Sales'] = df['Sales'].astype(float)
df['Quantity'] = df['Quantity'].astype(int)
df['Discount'] = df['Discount'].astype(float)
df['Profit'] = df['Profit'].astype(float)
df['Order Priority'] = df['Order Priority'].astype('category')
df['Payment Method'] = df['Payment Method'].astype('category')
df['Shipping Cost'] = df['Shipping Cost'].astype(float)
df['Customer Tenure'] = df['Customer Tenure'].astype(str)
df['Marketing Channel'] = df['Marketing Channel'].astype('category')
df['Product Rating'] = df['Product Rating'].astype(float)
df['Return Status'] = df['Return Status'].astype('category')
df
```

### **Remove Unnecessary Spaces**

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
df = df.apply(lambda x: x.str.strip() if x.dtype == 'category' else x)
```

# **Adding New Columns**

## **Shipping Delay**

```python
df['Shipping Delay'] = df.apply(lambda row: (row['Ship Date'] - row['Order Date']).days, axis=1)
df['Shipping Delay'] = df['Shipping Delay'].astype(int)
df
```

### **Profit Margin**

```python
df['Profit Margin'] = df.apply(lambda row: (row['Profit'] / row['Sales']) * 100, axis=1)
df['Profit Margin'] = df['Profit Margin'].astype(float)
df
```

```python
def Discount_Reason(Discount) :
                              if Discount == 0:
                                return 'No Discount'
                              elif Discount <=0.2:
                                return 'Promo'
                              elif Discount <= 0.4:
                                return 'Bulk Deal'
                              else:
                                return 'Clearance'

df['Discount_Reason'] = df['Discount'].apply(Discount_Reason)
df['Discount_Reason']
```

```python
df['Discount_Reason'] = df['Discount_Reason'].astype('category')
```
